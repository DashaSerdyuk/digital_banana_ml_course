```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
train_all = fetch_20newsgroups(subset='train')
simple_dataset = fetch_20newsgroups(
    subset='train', 
    categories=['sci.space', 'soc.religion.christian', 'comp.sys.mac.hardware'])

word_clusters = {}
```

```python
vectorizer = TfidfVectorizer(max_df=500, min_df=10)
matrix = vectorizer.fit_transform(simple_dataset.data)
```

```python
model = KMeans(n_clusters=3, random_state=1)
labels = model.fit_predict(matrix.toarray())
```

```python
def make_word_clusters(corpus = simple_dataset.data):
    print('Making word clusters...')
    for labelnumber in range(0,len(labels)):
        if labels[labelnumber] not in word_clusters:
            word_clusters[labels[labelnumber]] = []
        word_clusters[labels[labelnumber]].append(corpus[labelnumber])
    print('Word clusters have been created.')
```

```python
make_word_clusters(simple_dataset.data)
```

```python
real_themes = {1:"religion", 2: "space", 0: "hardware"}
for cluster_number in real_themes.keys():
    print("\n{1} {0} {1}\n".format(real_themes[cluster_number], '#'*15))
    for i in [0, 10, 20]:
        print("\n" +word_clusters[cluster_number][i][:500] + "\n")
        print('-'*45)
```
