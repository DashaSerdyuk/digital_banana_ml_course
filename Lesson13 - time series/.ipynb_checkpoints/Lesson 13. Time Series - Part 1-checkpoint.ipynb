2) В самом простом случае - это только целевая переменная, без признаков. Признаки требуется придумывать самим - либо извне, либо из самого таргета (на основе предыдущих значений)

### Сезонность и тренд

***Сезонность*** - это корреляция значений ряда с периодичностью в какой-то срок (неделя, месяц, год)

Например, выше у ряда есть годовая сезонность

***Тренд*** - это долговременная тенденция временного ряда. Например, рост объема рынка - если взглянуть на первые 8 лет продаж, то можно заметить, что они из года в год увеличиваются

```python
plt.plot(dataset['sales'][:96])
```

Как можно проверить сезонность? Пользуемся **автокорреляцией** - корреляция значений временно ряда с предыдущими значениями

```python
autocors = []

for i in range(1, 25):
    corr_coef = np.corrcoef(dataset['sales'][i:], dataset['sales'][:-i])[0][1]
    autocors.append(corr_coef)
```

```python
plt.plot(autocors)
```

## Простые модели - экспоненциальное сглаживание

<img src='exp_weight.png'>

```python
Y_exps = [dataset['sales'][0]]
alpha = 0.5

for i in range(1, len(dataset['sales'])):
    Y_exps.append(alpha * dataset['sales'][i] + (1 - alpha) * Y_exps[i-1])
```

```python
plt.plot(dataset['sales'])
plt.plot(Y_exps)
```

Предсказываем:

```python
alpha * dataset['sales'].values[-1] + (1 - alpha) * Y_exps[-1]
```
